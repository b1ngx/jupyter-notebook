### 爬虫

urllib2 和 BeautifulSoup 库的使用

```python
import urllib2
from bs4 import BeautifulSoup

url = 'https://mp.weixin.qq.com/s?__biz=MzA3NDM0ODQwMw==&mid=2649827536&idx=1&sn=a837a8c87ee7aa3f544a43013a305a07'
user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.100 Safari/537.36'


headers = {'User-Agent': user_agent}
request = urllib2.Request(url, headers = headers)

response = urllib2.urlopen(request)


page = response.read()

# 解析器可选 "html.parser" "lxml" "html5lib" "xml"
soup = BeautifulSoup(page, "lxml")

# 格式化输出
#print soup.prettify()

# 压缩输出 unicode 和 str
#unicode(soup)
#str(soup)

# 过滤
title = soup.find(id = 'activity-name').string.strip()
author = soup.find_all("em", class_="rich_media_meta rich_media_meta_text")[1].string
name = soup.find(id = 'post-user').string.strip()
#content = soup.find(id = 'js_content')

print author
print title
print name
#print content
```
